Import Libraries

```python
import pandas as pd 
import plotly.express as px
import pytz
from datetime import datetime
```

Load Dataset

```python
apps_df=pd.read_csv('Play Store Data.csv')
```

Data Cleaning (Installs Column)

```python
# Remove '+' and ',' from Installs
apps_df['Installs'] = apps_df['Installs'].str.replace('+', '', regex=False)
apps_df['Installs'] = apps_df['Installs'].str.replace(',', '', regex=False)

# Convert to numeric
apps_df['Installs'] = pd.to_numeric(apps_df['Installs'], errors='coerce')
```

Remove Categories Starting With A, C, G, S

```python
apps_df_filtered = apps_df[apps_df['Category'].str.startswith(('A', 'C', 'G', 'S'))]
```

Group by Category & Sum Installs

```python
category_installs = apps_df_filtered.groupby('Category')['Installs'].sum().reset_index()
```

Get Top 5 Categories

```python
top5 = category_installs.sort_values(by='Installs', ascending=False).head(5)
top5
```

Highlight Categories with Installs > 1 Million

```python
top5['Highlight'] = top5['Installs'] > 1000000
top5['Highlight']
```

Assign Countries for Choropleth

```python
#Since dataset contains no country column, assign the ISO country code
top5['ISO_Code'] = ['USA', 'IND', 'BRA', 'GBR', 'CAN']
```

Create Choropleth Map (Plotly)

```python
fig = px.choropleth(
    top5,
    locations="ISO_Code",
    locationmode="ISO-3",
    color="Installs",
    hover_name="Category",
    title="Top 5 App Categories - Global Installs",
    color_continuous_scale="Viridis"
)
fig.show()
```

Time Restriction Logic (6PM – 8PM IST)

```python
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist)
current_hour = current_time.hour

if 18 <= current_hour < 20:
    fig.show()
else:
    print("Choropleth map available only between 6 PM and 8 PM IST.")
```
